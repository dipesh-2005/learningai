```python
#copying the array
a=np.array([1,2,3])
b=a
print(b)
b[0]=100
print(b)
print(a)# when we change the content of b, a also changes because both are pointing to same memory location. To avoid this we use copy() function.
```

```python
#using copy function
a=np.array([2,3,4])
b=a.copy()
print(b) 
b[0]=100
print(b)
print(a)# now it doestnot change the original array a.
```

### linear algebra

```python
a=np.ones((3,3))
b=np.full((3,6),4)
np.matmul(a,b)
 
```

```python
#determinant of a matrix
c=np.identity(3)
np.linalg.det(c)
```

```python
#reorganizing an array
before=np.array([[1,2,3,4,1],[1,2,3,4,5]])
print(before)
after=before.reshape((5,2))
print(after)
```

```python
#vertically stacking vectors
v1=np.array([1,2,3,4])
v2=np.array([5,6,7,8])  
new=np.vstack([v1,v2])
print(new)
```
