Basic Plotting

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
plt.plot(x, y)
plt.title('My first plot', fontdict={'fontname':'Comic Sans MS', 'fontsize':20})
plt.xlabel('x axis label', fontdict={'fontname':'Arial', 'fontsize':15})
plt.ylabel('y axis label')
plt.show()
```
