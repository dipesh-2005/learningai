Basic Plotting

```python
x=[1,2,3,4,5]#x values
y=[1,4,9,16,25]#y values which are squares of x values
plt.plot(x, y)#to plot x and y values
plt.title('My first plot', fontdict={'fontname':'Comic Sans MS', 'fontsize':20})#to set title of the plot with font name and size
plt.xlabel('x axis label', fontdict={'fontname':'Arial', 'fontsize':15})#to set x axis label with font name and size
plt.ylabel('y axis label')#to set y axis label
plt.grid()#to show grid lines
plt.show()#to display the plot

```

```python
data={
    "salary" : [25000,30000,35000,40000,25000,16000,17000,80000,19000,100000,110000,20000,30000,18000,15000,16000,17000,18000,19000,45000]
}
df=pd.DataFrame(data)
df.head(5)
```

```python
df.shape
```

```python
#line plot
plt.plot(df['salary'] , color='red' , marker='o', linestyle='--')#to plot salary values with red color and circular markers can do linestyle =':' to make dotted lines.
plt.grid() 
plt.show()
```

USING HISTOGRAM TO VISUALISE THE DATA

```python
plt.hist(df['salary'], bins=5, color='blue', edgecolor='black')#to plot histogram of salary values with 5 bins and blue color
plt.title('Salary Distribution')#to set title of the plot
plt.xlabel('Salary')#to set x axis label
plt.ylabel('Frequency')#to set y axis label    
plt.show()
```

using boxplot

```python
plt.boxplot(df['salary'])
plt.show()
#there are outliers which are values that are significantly different from the rest of the data. In this case, the outliers are the salary values that are much higher than the majority of the data points, such as 80000, 100000, and 110000. These values can skew the distribution and affect the overall analysis of the data. they are represented as individual points outside the whiskers of the box plot.
```

USING PIE CHART

```python
count=df['dept'].value_counts()
count
```

```python
plt.pie(count, labels=count.index, autopct='%1.1f%%', explode=[0.2,0.1,0.1,0.1], shadow=True) #to plot pie chart of department distribution with labels, percentage, explode to highlight HR department and shadow for 3D effect. explode is a list of values that specify how much to "explode" each slice of the pie chart. In this case, the HR department is exploded by 0.2, while the other departments are exploded by 0.1. autopct is a string format that specifies how to display the percentage values on the pie chart. In this case, '%1.1f%%' means to display the percentage with one decimal place followed by a percent sign. labels is set to count.index to use the department names as labels for the pie chart. shadow=True adds a shadow effect to the pie chart for better visualization.
plt.axis('equal')#to make the pie chart circular instead of elliptical 
plt.title('Department Distribution')#to set title of the plot
```

COUNTPLOT

```python
plt.bar(count.index, count.values, color=['red', 'blue', 'green', 'orange'])#to plot bar chart of department distribution with department names on x axis and counts on y axis with different colors for each bar\
plt.title('Department Distribution')#to set title of the plot
plt.show()
```

```python
df['Experience']=[2,5,7,10,3,4,6,8,1,12,15,20,25,30,35,40,45,50,55,60]
df.head(5)

```

```python
#bubble plot
plt.scatter(df['Age'], df['salary'], s=df['Experience']*10, color='cyan', alpha=0.5, edgecolor='black')#to plot bubble chart of age vs salary with bubble size representing experience, cyan color, 50% transparency and black edge color.
plt.title('Age vs Salary with Experience')#to set title of the plot
plt.show()
```

```python
#two numerical and  1 categorical column
plt.scatter(df['Age'], df['salary'], c = df['dept'] .map({'IT': 'red', 'HR': 'blue', 'Finance': 'green', 'Marketing': 'orange'}))#to plot scatter plot of age vs salary with color representing department using rainbow colormap
plt.xlabel('Age')#to set x axis label
plt.ylabel('Salary')#to set y axis label
plt.legend(['IT', 'HR', 'Finance', 'Marketing'])#to add legend to the plot to differentiate between departments based on color
plt.title('Age vs Salary by Department')#to set title of the plot
plt.show()
```

```python
color={
    "IT": 'red',
    "HR": 'blue',
    "Finance": 'green',
    "Marketing": 'orange'
}
for dept, color in color.items():
    df_dept = df[df['dept'] == dept]#to filter the dataframe for each department based on the department column
    plt.scatter(df_dept['Age'], df_dept['salary'], color=color, label=dept)#to plot scatter plot of age vs salary for each department with different colors and labels for each department
    plt.legend()#to add legend to the plot to differentiate between departments based on color
```

OBJECT ORIENTED API

```python
fig, axs=plt.subplots(1,3, figsize=(15,5) )#to create a figure with 1 row and 3 columns of subplots with a figure size of 10 inches by 5 inches. axs is a variable that holds the array of subplot axes created by plt.subplots(). Each element in axs corresponds to a subplot, allowing you to customize each subplot individually.

#line plot
axs[0].plot(df['Age'], df['salary'], color='red', marker='o', linestyle='--', linewidth=2, markersize=2)#to plot line plot of age vs salary on the first subplot with red color, circular markers and dashed lines
axs[0].grid()
axs[0].set_title('Line plot')#to set title of the first subplot
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Salary')

#histogram
axs[1].hist(df['salary'], bins=5, color='blue', edgecolor='black')#to plot histogram of salary values on the second subplot with 5 bins and blue color
axs[1].set_title('Histogram') #to set title of the second subplot
axs[1].set_xlabel('Salary')
axs[1].set_ylabel('Frequency')

#boxplot
axs[2].boxplot(df['salary'])#to plot boxplot of salary values on the third subplot
axs[2].set_title('Boxplot')#to set title of the third subplot
axs[2].set_xlabel('Salary')#to set x axis label for the third subplot
axs[2].set_ylabel('Salary')#to set y axis label for the third subplot


# to save the plot as an image file
# plt.savefig('age_salary_plot.png')#to save the current figure as an image file named 'age_salary_plot.png' in the current working directory. The savefig() function is used to save the figure, and the filename is specified as a parameter. You can also specify the file format (e.g., .jpg, .pdf) by changing the file extension in the filename. cana lso specify the paths i want to save the picture in a particular folder by providing the full path like 'C:/AIML/Matplotlib/age_salary_plot.png'
plt.savefig('C:/AIML/Data/age_salary_plot.png')


plt.show()

```
