Loading Necessary Libraries

```python
import matplotlib.pyplot as plt
import numpy as np   
import pandas as pd
```

Basic Plotting

```python
x=[1,2,3,4,5]#x values
y=[1,4,9,16,25]#y values which are squares of x values
plt.plot(x, y)#to plot x and y values
plt.title('My first plot', fontdict={'fontname':'Comic Sans MS', 'fontsize':20})#to set title of the plot with font name and size
plt.xlabel('x axis label', fontdict={'fontname':'Arial', 'fontsize':15})#to set x axis label with font name and size
plt.ylabel('y axis label')#to set y axis label
plt.grid()#to show grid lines
plt.show()#to display the plot

```

LINE PLOT

####

```python
data={
    "salary" : [25000,30000,35000,40000,25000,16000,17000,80000,19000,100000,110000,20000,30000,18000,15000,16000,17000,18000,19000,45000]
}
df=pd.DataFrame(data)
df.head(5)
```

```python
df.shape
```

```python
#line plot
plt.plot(df['salary'] , color='red' , marker='o', linestyle='--')#to plot salary values with red color and circular markers can do linestyle =':' to make dotted lines.
plt.grid() 
plt.show()
```

USING HISTOGRAM TO VISUALISE THE DATA

```python
plt.hist(df['salary'], bins=5, color='blue', edgecolor='black')#to plot histogram of salary values with 5 bins and blue color
plt.title('Salary Distribution')#to set title of the plot
plt.xlabel('Salary')#to set x axis label
plt.ylabel('Frequency')#to set y axis label    
plt.show()
```

using boxplot

```python
plt.boxplot(df['salary'])
plt.show()
#there are outliers which are values that are significantly different from the rest of the data. In this case, the outliers are the salary values that are much higher than the majority of the data points, such as 80000, 100000, and 110000. These values can skew the distribution and affect the overall analysis of the data. they are represented as individual points outside the whiskers of the box plot.
```

```python
df['dept']=['IT','HR','Finance','Marketing','IT','HR','Finance','Marketing','IT','HR','Finance','Marketing','IT','HR','Finance','Marketing','IT','HR','Finance','Marketing']
df.head(5)
```

USING PIE CHART

```python
count=df['dept'].value_counts()
count
```

```python
plt.pie(count, labels=count.index, autopct='%1.1f%%', explode=[0.2,0.1,0.1,0.1], shadow=True) #to plot pie chart of department distribution with labels, percentage, explode to highlight HR department and shadow for 3D effect. explode is a list of values that specify how much to "explode" each slice of the pie chart. In this case, the HR department is exploded by 0.2, while the other departments are exploded by 0.1. autopct is a string format that specifies how to display the percentage values on the pie chart. In this case, '%1.1f%%' means to display the percentage with one decimal place followed by a percent sign. labels is set to count.index to use the department names as labels for the pie chart. shadow=True adds a shadow effect to the pie chart for better visualization.
plt.axis('equal')#to make the pie chart circular instead of elliptical 
plt.title('Department Distribution')#to set title of the plot
```

COUNTPLOT

```python
plt.bar(count.index, count.values, color=['red', 'blue', 'green', 'orange'])#to plot bar chart of department distribution with department names on x axis and counts on y axis with different colors for each bar\
plt.title('Department Distribution')#to set title of the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Needed for 3D projection
import numpy as np

# Sample data for 3D scatter plot 
x = np.random.rand(50)#to generate 50 random numbers between 0 and 1 for x values
y = np.random.rand(50)#to generate 50 random numbers between 0 and 1 for y values
z = np.random.rand(50)#to generate 50 random numbers between 0 and 1 for z values

# Create figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') #to create a 3D subplot  

# Create 3D scatter plot
ax.scatter(x, y, z)

# Labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
ax.set_title('Simple 3D Scatter Plot')

plt.show()
```
